Une autre mise à jour récente supprime la nécessité d’activer les compétences avant de les utiliser. Vous pouvez toujours activer ou désactiver manuellement les compétences, mais simplement dire “Ouvrir [skill invocation]”trouvera et activera une compétence. Mais si cette compétence vous oblige à vous connecter à votre compte, par exemple avec les commandes Starbucks Reorder ou Domino, vous devrez toujours ouvrir l’application Alexa ou aller sur alexa.amazon.com, ouvrez le menu compétences et connectez-vous à votre compte pour autoriser la connexion.

# Commandes vocales pour Alexa et les machines à café Home Connect

«Alexa, démarre la machine à café Home Connect» ou «Alexa, dis à la machine à café Home Connect de démarrer». Alexa entame le dialogue immédiatement, et vous guide tout au long des étapes.

Si vous maîtrisez déjà la compétence, essayez de faire une demande plus complexe; par exemple: «Alexa, dis à la machine à café Home Connect de faire un cappuccino.». Ce raccourci vous permet de sélectionner votre café préféré rapidement et directement en une seule phrase. Pour cela, la machine à café utilise les derniers réglages utilisés pour la contenance et la force.

Si vous avez une préférence personnelle pour le café, donnez directement toutes les indications: «Alexa, dis à la machine à café Home Connect de faire un grand cappuccino très serré.». Si Alexa ne comprends pas correctement tous

les mots, l'Assistant vient automatiquement à votre rescousse.

Il se peut également que la machine à café soit en veille au moment de votre demande. Dans ce cas, la machine à café démarre automatiquement et lance le cycle de rinçage. Patientez quelques minutes jusqu'à la fin du cycle et recommencez.

Voici quelques conseils: si vous ne savez pas exactement quoi dire pendant un dialogue, attendez quelques secondes, Alexa fera automatiquement quelques suggestions. Vous pouvez mettre fin au dialogue quand vous le souhaitez en disant «Interrompre» ou «Arrêter».
