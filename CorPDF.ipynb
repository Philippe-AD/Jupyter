```
#a function to pre process Coraline by Neil Gaiman

def preprocess_coraline(book):
  '''
  param book: url od a PDF book file
  '''
  output = ""
  data = open(book, 'rb')
  data = PyPDF2.PdfFileReader(book)
  npages = data.getNumPages()
  for i in range(npages):
    page_i = data.getPage(i).extractText()
    output += page_i

  output = output[1227:]
  output = output.lower()
  
  processed = word_tokenize(output)
  print('Coraline database includes {} tokens, and {} unique tokens after editing'.format(len(processed), len(set(processed)))) 
  return processed

coraline = preprocess_coraline(coraline_url) 
print(coraline)
output = coraline 
```

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(binary=True)
corpus = [
'This movie is very scary and long',
'This movie is not scary and is slow',
'This movie is spooky and good',
 ]
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names()

# words in vocablary and order in which feature are arranged
output : ['and', 'good', 'is', 'long', 'movie', 'not',
          'scary', 'slow', 'spooky', 'this', 'very']

X.toarray()
#Document represented in BOW binary Vector format
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(binary=True)
corpus = [
'This movie is very scary and long',
'This movie is not scary and is slow',
'This movie is spooky and good',
 ]
X = vectorizer.fit_transform(corpus)
vectorizer.get_feature_names()

# words in vocablary and order in which feature are arranged
output : ['and', 'good', 'is', 'long', 'movie', 'not',
          'scary', 'slow', 'spooky', 'this', 'very']

X.toarray()
```

```
from keras.preprocessing.text import Tokenizer

t  = Tokenizer()
# Defining 4 document lists
fit_text = ['Machine Learning Knowledge',
	          'Machine Learning',
            'Deep Learning',
            'Artificial Intelligence']

t.fit_on_texts(fit_text)
print("The document count",t.document_count)
print("The count of words",t.word_counts)
print("The word index",t.word_index)
print("The word docs",t.word_docs)
```

```
t = Tokenizer()

test_text = ['Machine Learning Knowledge',
	          'Machine Learning',
            'Deep Learning',
            'Artificial Intelligence']

t.fit_on_texts(test_text)

sequences = t.texts_to_sequences(test_text)

print("The sequences generated from text are : ",sequences)
```
